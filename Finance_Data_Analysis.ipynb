In this study I present descriptive analysis using bar charts for finance data. Specifically, I present the analysis for investment avenues, stock market paarticipation, factors affecting investment, etc.  I report the interpretation of the results below each bar chart. I also explore the differences between male and female respondents in their responses. Finally, for several variables I investigate the impact of age on responses.

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Reading the data into dataframe
df = pd.read_csv('Finance_data.csv')
df.info()
```

```python
#  Bar chart for gender
plt.figure(figsize=(8, 5))
sns.countplot(x=df['gender'])
plt.title('Count of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

The above bar chart shows that 15 of respondents are female while 25 of them are male. This result suggests that majority of respondents are male.

```python
# Histogram of Age
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='skyblue', ax=ax)
ax.set_title('Age Distribution')
ax.set_xlabel('Age') # Add an x-label for clarity
ax.set_ylabel('Frequency') # Add a y-label for clarity
plt.show()
```

The above graph shows that most of our respondents were in the ages 24-31. Surprisingly, there are not any respondent older than 35.

```python
# Bar chart for Investment_Avenues
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Investment_Avenues'], palette='summer',linewidth=3)
plt.title('Count of Investment Avenues')
plt.xlabel('Investment Avenue')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

The above bar chart shows that most of the respondents (37 out of 40 of them) use investment avenues.

```python
# Bar chart for Stock_Marktet Participation
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Stock_Marktet'], palette='hot')
plt.title('Count of Stock Market Participation')
plt.xlabel('Stock Market Participation')
plt.ylabel('Count')
plt.show()
```

The above bar chart shows that most of the respondents (35 out of 40 of them) participate in stock market.

```python
# Bar chart for the interaction of gender with Investment_Avenue and Stock_Market
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(x=df['gender'],hue=df['Investment_Avenues'],palette='summer',linewidth=3,edgecolor='white')
plt.title('INVESTMENT AVENUE')
plt.subplot(1, 2, 2)
sns.countplot(x=df['gender'],hue=df['Stock_Marktet'],palette='hot',linewidth=3,edgecolor='white')
plt.title('STOCK MARKET')
plt.show()
```

The above bar chart on the left shows that while all 25 of the male respondents use investment avenues, three of female respondents dont use them. Hence, the use of investment avenues is more common among males. The bar chart on the right shows that only one out of 25 male respondents dont participate stock market. However, four out of 15 female respondents dont participate in stock market. These findings suggest that stock market participation is more common among male respondents.

```python
# Bar chart for investment objective
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Objective'], palette='viridis')
plt.title('Count of Investment Objectives')
plt.xlabel('Objective')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The above bar chart shows that the most important investment objective is capital appreciation (26 out of 40 respondents) followed by growth. Only three respondents consider income as investment objective suggesting that it is the weakest investment objective.

```python
# Bar chart for the interaction of gender with investment objective
sns.countplot(x=df['gender'],hue=df['Objective'],palette='viridis',linewidth=2)
plt.title('Gender')
plt.show()
```

The above bar chart shows that capital appreciation is the most important investment objective for both male and female respondents. Two female respondent consider income as investment objective while only one male respondent considers it as investment objective. Hence, females are more likely to consider income as investment objective.

```python
# Relationship between age and Investing Objective
plt.figure(figsize=(16,6))
sns.pointplot(x="Objective",y='age',data=df,linestyles="--",capsize=.3,color='Red')
plt.show()
```

The above plot shows that respondents in middle ages choose capital appreciation or income as their investment objective and as respondents get older growth seems to be more important.

```python
# Factors affecting Investing
sns.countplot(x=df['Factor'],palette='coolwarm',linewidth=2)
plt.title('Factors affecting Investing')
plt.xlabel('Factors')
plt.ylabel('Count')
plt.show()
```

The above bar chart shows that the most important factor affecting investing is returns followed by risk. Only one respondent out of 40 responds with locking period factors as factor affecting investing.

```python
# Bar chart for the interaction of gender with Factor
sns.countplot(x=df['gender'],hue=df['Factor'],palette='coolwarm',linewidth=2)
plt.title('Gender')
plt.show()
```

The obve bar chart shows that returns is the most important factor affecting investing for both male and female respondents followed by risk for both groups. While locking period is selected by one female repondent, it is not selected by any male respondent.

The above bar chart shows that wealth creation is the most important investment purpose while saving for future is the second important purpose.

```python
# Relationship between age and Investing Purpose
plt.figure(figsize=(16,6))
sns.pointplot(x="Purpose",y='age',data=df,linestyles="--",capsize=.3,color='Red')
plt.show()
```

The above plot shows that as people grow older they are more inclined towards Returns and people at young age look more for wealth creation. Hence, age seem to affect investment purpose.

```python
# Bar chart for Duration
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Duration'], palette='plasma')
plt.title('Count of Investment Duration')
plt.xlabel('Duration')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The above bar chart shows that most of the respondents invest for 1-5 years and it is very rare to invest more than 5 years or less than 1 year.

```python
# Bar chart  for Invest_Monitor
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Invest_Monitor'], palette='seismic')
plt.title('Count of Investment Monitoring Frequency')
plt.xlabel('Investment Monitoring Frequency')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The above bar chart shows that most of the respondents monitor their investments monthly and it is very rare to monitor daily. There is is no respondent monitoring weekly in our sample.

```python
# The interaction of gender with Duration and Invest_Monitor
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(hue=df['gender'],x=df['Duration'],palette='viridis',linewidth=2,edgecolor='black')
plt.title('DURATION')
plt.subplot(1, 2, 2)
sns.countplot(hue=df['gender'],x=df['Invest_Monitor'],palette='seismic',linewidth=2,edgecolor='black')
plt.title('INVESTMENT MONITORING')
plt.show()
```

The above bar chart on the left shows that the most important investment duration for both male and female respondents is 1-5 years. The bar chart on the right shows that both male and female respondents are more likely to monitor their investments monthly.

```python
# Relationship between age and Duration
plt.figure(figsize=(16,6))
sns.pointplot(x="Duration",y='age',data=df,linestyles="--",capsize=.3,color='Red')
plt.show()
```

The above plot shows that investment duration for respondents at younger ages is either more than 5 years or less than 1 year, while the duration for older ages is 1-5 years.

```python
# Bar chart of Avennue
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Avenue'], palette='summer',linewidth=3)
plt.title('Count of Avenue')
plt.xlabel('Avenue')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

The above bar chart shows that most important investment avenue is mutual fund while the least important avenue is public provident fund.

```python
# Bar chart for the interaction of gender with Avenue
sns.countplot(x=df['gender'],hue=df['Avenue'],palette='summer',linewidth=2)
plt.title('Gender')
plt.show()
```

The above bar chart shows that mutual fund is the most important investment avenue for both female and male respondents followed by equity investment. Public provident fund is the least important avenue for both groups.

```python
# Relationship between age and Avenue
plt.figure(figsize=(16,6))
sns.pointplot(x="Avenue",y='age',data=df,linestyles="--",capsize=.3,color='Red')
plt.show()
```

The above plot shows that investment avenue for respondents at young ages is public provident fund while respondents older ages choose other investment avenues.

```python
# Bar chart for Reason_Equity
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Reason_Equity'], palette='viridis')
plt.title('Count of Reasons for Investing in Equity')
plt.xlabel('Reason for Equity Investment')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The above bar chart shows that the most imoportant reason to invest in equity is capital appreciation while liquidity is the least important reason.

The above bar charts shows that capital appreciation is the most important reason for investing in equity for both female and male respondents and liquidity is the least important for both groups.

```python
# Bar chart for Reason_Mutual Fund
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Reason_Mutual'], palette='viridis')
plt.title('Count of Reasons for Investing in Mutual Funds')
plt.xlabel('Reason for Mutual Fund Investment')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The above bar chart shows that better returns is the most important reason to invest in mutual funds while tax benefits is the least important reason.

The above bar charts shows that while better returns is the most important reason to invest in mutual funds for both female and male respondents, fund diversifation is much more important reason for males compared to females. Tax benefits is the least important reason for both groups.

```python
# Bar chart for Reason_Bonds
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Reason_Bonds'], palette='viridis')
plt.title('Count of Reasons for Investing in Bonds')
plt.xlabel('Reason for Bond Investment')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The above bar chart shows that assured returns is the most important reasonto invest in bonds followed by bonds being safe investment, while tax incentives is the least important reson.

The above bar chart shows that assured returns is the most important reason to invest in bonds for both female and male respondents while only male respondents cite tax incentives as reason to invest in bonds.

```python
# Bar chart for Reason_FD
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Reason_FD'], palette='viridis')
plt.title('Count of Reasons for Investing in Fixed Deposits')
plt.xlabel('Reason for Fixed Deposit Investment')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The above bar chart shows that fixed returns and being risk free are the two most important reasons to invest in fixed deposit investment while high interest rates is the least important reason.
