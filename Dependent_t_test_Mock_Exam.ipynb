```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro, probplot, ttest_rel
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Dependent Ttest - Mock Exam.csv')

df
```

```python
before = df['Before']
after = df['After']

differences = after - before

differences
```

```python
differences.describe()
```

```python
# Q-Q plot for visual inspection
plt.figure(figsize=(6, 6))
probplot(differences, dist="norm", plot=plt)
plt.title('Q-Q Plot of Differences')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.hist(differences, bins=10, edgecolor='black')
plt.title('Histogram of Differences')
plt.xlabel('Differences')
plt.ylabel('Frequency')
plt.show()
```

```python
# Perform Shapiro-Wilk test for normality
shapiro_test = shapiro(differences)
print("Shapiro-Wilk test statistic:", shapiro_test.statistic)
print("Shapiro-Wilk p-value:", shapiro_test.pvalue)
```

```python
# Perform paired t-test
result = ttest_rel(before, after)

result
```

```python
# nonparametric counterpart

from scipy.stats import wilcoxon

# Perform Wilcoxon signed-rank test
result = wilcoxon(differences)

print(result)
```
